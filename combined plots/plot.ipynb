```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_csv_and_plot_same_metric(csv_files, metric, save_path= None):
    data_dicts = []

    # Read each CSV file and store data in a dictionary
    for csv_file in csv_files:
        data_dict = {}
        df = pd.read_csv(csv_file)
        epochs = df['Epoch']
        data_dict[csv_file] = {epoch: value for epoch, value in zip(epochs, df[metric])}
        data_dicts.append(data_dict)

    # Plot data
    for data_dict in data_dicts:
        for label, values in data_dict.items():
            plt.plot(list(values.keys()), list(values.values()), label=label)

    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.title(f'{metric} across epochs')
    plt.legend()
    plt.show()
    if save_path:
        plt.savefig(save_path)
    else:
        plt.show()  
```

```python
csv_files= ['lstm-lstm.csv', 'lstm-lstm-attn.csv', 'bert-frozen.csv','bert-finetune.csv']
```

```python
metric_to_plot = 'Train Loss'
read_csv_and_plot_same_metric(csv_files, metric_to_plot)

metric_to_plot = 'Val Loss'
read_csv_and_plot_same_metric(csv_files, metric_to_plot)
```
